```python
# Fib
import time
def fib(n,d):
    if n in d:
        return d[n]
    else:
        d[n] = fib(n-1,d)+fib(n-2,d)
        return d[n]
```

```python
start = time.time()
d={0:1,1:1}
fib(12,d)
end = time.time()
end-start
```

```python
start = time.time()
fib(36,d)
end = time.time()
end-start
```
